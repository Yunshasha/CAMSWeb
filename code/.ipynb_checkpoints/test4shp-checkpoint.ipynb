```python
import geopandas
from rasterio import features
from affine import Affine
import xarray
import numpy as np

def transform_from_latlon(lat, lon):
    lat = np.asarray(lat)
    lon = np.asarray(lon)
    trans = Affine.translation(lon[0], lat[0])
    scale = Affine.scale(lon[1] - lon[0], lat[1] - lat[0])
    return trans * scale

def rasterize(shapes, coords, fill=np.nan, **kwargs):
    """Rasterize a list of (geometry, fill_value) tuples onto the given
    xray coordinates. This only works for 1d latitude and longitudewoe_name
    arrays.
    """
    transform = transform_from_latlon(coords['latitude'], coords['longitude'])
    out_shape = (len(coords['latitude']), len(coords['longitude']))
    raster = features.rasterize(shapes, out_shape=out_shape,
                                fill=fill, transform=transform,
                                dtype=float, **kwargs)
    return xarray.DataArray(raster, coords=coords, dims=('latitude', 'longitude'))
```

```python
# this shapefile is from natural earth data
# http://www.naturalearthdata.com/downloads/10m-cultural-vectors/10m-admin-1-states-provinces/

states = geopandas.read_file('data/ne_10m_admin_1_states_provinces_lakes')
states.plot()
```

```python
it_states = states.query("admin == 'Italy'").reset_index(drop=True)
it_states.plot()
```

```python
state_ids = {k: i for i, k in enumerate(it_states.woe_name)}
shapes = [(shape, n) for n, shape in enumerate(it_states.geometry)]

state_ids['Milano']
```

```python
myds = xarray.open_dataset("data/download.nc")
myds.rio.write_crs("epsg:4326", inplace=True)
myds = myds.assign_coords(longitude=(((myds.longitude + 180) % 360) - 180)).sortby('longitude')
myds
```

```python

ds = xarray.Dataset(coords={'longitude': myds.longitude,
                          'latitude':myds.latitude})
ds
```

```python
ds['states'] = rasterize(shapes, {'longitude':ds.longitude,
                                     'latitude':ds.latitude})
ds
```
